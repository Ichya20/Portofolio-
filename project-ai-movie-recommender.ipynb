```python
movies=movies[[ 'id','genres', 'keywords', 'original_title', 'overview', 'cast', 'crew']] 
#taking only the feature required for movie recommendation
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies['genres'][0] 
```

```python
import ast
def convert(obj):
    tags=[]
    for i in ast.literal_eval(obj): 
        tags.append(i["name"])
    return tags
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
import ast
def convert3(obj):
    tags=[]
    iterator=0;
    for i in ast.literal_eval(obj):
        if iterator!=3:   #top three in the cast
            tags.append(i["name"])
            iterator+=1;
        else:
            break
    return tags
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(2)
```

```python
movies['crew'][0]
```

```python
def director(obj): #director extraction from crew
    tags=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':   #Since only 1 director in movie
            tags.append(i["name"]) 
            break
    return tags
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
movies.head(2)
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
#key words lets make them a single example Sam Worthington as SamWorthington since Sam Worthington will be one tag
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['id','original_title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(2)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stemming(text):
    an=[]
    for i in text.split():
        an.append(ps.stem(i))
    return " ".join(an)
```

```python
new_df['tags']=new_df['tags'].apply(stemming)
```

```python
new_df.head(2)
```

```python
#Text Vecterization using Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
new_df.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommender(movie):
    movie_index=new_df[new_df['original_title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] #top five matching movies
    
    for i in movies_list:
        print(new_df.iloc[i[0]].original_title)
```

```python
recommender('Avatar')
```

```python
recommender('The Dark Knight Rises')
```

```python
recommender('The Avengers')
```
